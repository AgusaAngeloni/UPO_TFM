```python
%%R
# =================== INSTALACIÓN DE LIBRERIAS ============================================
# Instalar y cargar HiPathia
library(recount3) 
library(edgeR)
library(SummarizedExperiment)
library(hipathia) 
```

```python
%%R
# =================== OBTENCIÓN DE DATOS ==================================================
####-- BLOOD
gtex_blood <- subset(available_projects(), file_source == "gtex" & 
                           project =="BLOOD")
rseBLOOD <- create_rse(gtex_blood)

####-- THYROID
gtex_thyr <- subset(available_projects(), file_source == "gtex" & 
                          project =="THYROID")
rseTHYR<- create_rse(gtex_thyr)
```

```python
%%R
#=================================== OTROS TEJIDOS =================================
####-- OVARY
gtex_ovary <- subset(available_projects(), file_source == "gtex" & 
                          project =="OVARY")
rseOVARY<- create_rse(gtex_ovary)

####-- OVARY
gtex_lung <- subset(available_projects(), file_source == "gtex" & 
                          project =="LUNG")
rseLUNG<- create_rse(gtex_lung)
    
```

```python
%%R
gtex_projects <- subset(available_projects(), file_source == "gtex")
gtex_projects 
```

```python
%%R
# ============================= Limitación de los datos de interes =======================
## Limitación de los datos de interes
rseBLOOD <- rseBLOOD[, rseBLOOD$gtex.smtsd == "Whole Blood" & rseBLOOD$gtex.smafrze == "RNASEQ"]
rseTHYR <- rseTHYR[, rseTHYR$gtex.smafrze == "RNASEQ"]
```

```python
%%R
# ============================= OTROS TEJIDOS - Limitación de los datos de interes =========
rseLUNG <- rseLUNG[, rseLUNG$gtex.smafrze == "RNASEQ"]
rseOVARY <- rseOVARY[, rseOVARY$gtex.smafrze == "RNASEQ"]
```

```python
%%R
# ============================= CONTEOS CRUODOS Y METADATA ================================
## Obtención por separado de arrays con los conteos crudos de RNA-seq y los 
## datos clínicos y experimentales en metadata
countsBLOOD <- assay(rseBLOOD, "raw_counts")  #Matriz de conteos crudos
metadataBLOOD <- colData(rseBLOOD)            #df de datos clínicos y experimentales
####-- THYROID
countsTHYR <- assay(rseTHYR, "raw_counts")  #Matriz de conteos crudos
metadataTHYR <- colData(rseTHYR)            #df de datos clínicos y experimentales
```

```python
%%R
#=================================== OTROS TEJIDOS =================================
countsLUNG <- assay(rseLUNG, "raw_counts")  #Matriz de conteos crudos
metadataLUNG<- colData(rseLUNG)            #df de datos clínicos y experimentales
####-- THYROID
countsOVARY <- assay(rseOVARY, "raw_counts")  #Matriz de conteos crudos
metadataOVARY <- colData(rseOVARY)            #df de datos clínicos y experimentales
```

```python
%%R
### ========================  RE-FILTRADO  =================================
keepBLOOD <- rowSums(countsBLOOD > 0) >= 5  # Genes expresados en ≥5 muestras
countsBLOOD <- countsBLOOD[keepBLOOD, ] #De 63856 genes a 55424 genes
####-- THYROID
keepTHYR <- rowSums(countsTHYR > 0) >= 5  # Genes expresados en ≥5 muestras
countsTHYR <- countsTHYR[keepTHYR, ] #De 63856 genes a 55424 genes
```

```python
%%R
### ======================== OTROS TEJIDOS - PRE-FILTRADO ===============================
keepLUNG <- rowSums(countsLUNG > 0) >= 5  # Genes expresados en ≥5 muestras
countsLUNG <- countsLUNG[keepLUNG, ] #De 63856 genes a 55424 genes
```

```python
%%R
## ============================ NORMALIZACIÓN TMM EN DATOS CRUDOS =====================
    # Normalización TMM = Trimmed Mean of M-values:
    # método para corregir diferencias en la profundidad de secuenciación y 
    # en la composición global del RNA entre muestras.
    # Conversión counts crudos a CPM (Counts Per Million),
    # usando tamaños de librería ajustados con factores de normalización TMM 
    # (calcula lib.size * norm.factors)

dgeB <- DGEList(counts = countsBLOOD) # clase DGEList (Differential Gene Expression List)
dge_normB <- cpm(dgeB, 
                    normalized.lib.sizes = TRUE, 
                    log = FALSE)                              
                     
dge_normalizedB <- log2(dge_normB + 1) # transformación log2 para obtener valores continuos comparables 

####-- THYROID
dgeT <- DGEList(counts = countsTHYR) # clase DGEList (Differential Gene Expression List)
dge_normT <- cpm(dgeT, 
                    normalized.lib.sizes = TRUE, 
                    log = FALSE)                              
                     
dge_normalizedT <- log2(dge_normT + 1) 
```

```python
%%R

## ===================== OTROS TEJIDOS - NORMALIZACIÓN TMM EN DATOS CRUDOS ==================
####-- LUNG
dgeL <- DGEList(counts = countsLUNG) # clase DGEList (Differential Gene Expression List)
dge_normL <- cpm(dgeL, 
                    normalized.lib.sizes = TRUE, 
                    log = FALSE)                              
                     
dge_normalizedL <- log2(dge_normL + 1) 
```

```python
%%R
### ===================== CAMBIO DE ENSEMBL A ENTREZ (Hipathia) ===========================
data_entrezB <- translate_data(dge_normalizedB, "hsa")
      # translated ids = 23736 (0.43) 
      # untranslated ids = 31688 (0.57) 
      # multihit ids = 181 (0.0033) 
    ## De los 55424 genes por Ensembl, se reconocen 23638 con id Entrez en Hipathia
table(is.na(rownames(data_entrezB)))
    # IMPORTANTE: que no existan filas con id no reconocidos por Hipathia
```

```python
%%R
### =========================== CAMBIO DE ENSEMBL A ENTREZ (Hipathia) =====================
data_entrezT <- translate_data(dge_normalizedT, "hsa")
      # translated ids = 23736 (0.43) 
      # untranslated ids = 31688 (0.57) 
      # multihit ids = 181 (0.0033) 
    ## De los 55424 genes por Ensembl, se reconocen 23638 con id Entrez en Hipathia
table(is.na(rownames(data_entrezT)))
    # IMPORTANTE: que no existan filas con id no reconocidos por Hipathia
```

```python
%%R
### =========================== CAMBIO DE ENSEMBL A ENTREZ (Hipathia) =====================
data_entrezL <- translate_data(dge_normalizedL, "hsa")
      # translated ids = 23736 (0.43) 
      # untranslated ids = 31688 (0.57) 
      # multihit ids = 181 (0.0033) 
    ## De los 55424 genes por Ensembl, se reconocen 23638 con id Entrez en Hipathia
table(is.na(rownames(data_entrezL)))
    # IMPORTANTE: que no existan filas con id no reconocidos por Hipathia
```

```python
%%R
### ================================ NORMALIZACION a [0,1] (Hipathia) ======================
data_entrez_normB <- normalize_data(data_entrezB)
data_entrez_normT <- normalize_data(data_entrezT)
```

```python
%%R
### ================================ NORMALIZACION a [0,1] (Hipathia) ======================
data_entrez_normL <- normalize_data(data_entrezL)
```

```python
%%R
### ======================= OBJETO FINAL PARA HIPATHIA SummarizedEperiment ======================
data_infoT <- SummarizedExperiment(
    assays = list(counts = data_entrez_normT), #counts normalizados
     colData = DataFrame(
        tissue = colData(rseTHYR)$gtex.smts,
        subtissue = colData(rseTHYR)$gtex.smtsd
    )
)
```

```python
%%R
### 6 - OBJETO FINAL PARA HIPATHIA SummarizedEperiment
data_infoL <- SummarizedExperiment(
    assays = list(counts = data_entrez_normL), #counts normalizados
     colData = DataFrame(
        tissue = colData(rseLUNG)$gtex.smts,
        subtissue = colData(rseLUNG)$gtex.smtsd
    )
)
```

```python
%%R
### ================== DESCARGA PATHWAYS HUMANOS (Hipathia) =================================
pathways <- load_pathways("hsa")
        # Loaded 146 pathways
```

```python
%%R
    
### ======================= CALCULO DE ACTIVIDAD DE RUTAS =========================================
## HIPATHIA - result
results <- hipathia(data_infoT, pathways) # utiliza las subrutas efectoras
                                             # opción decompose=TRUE utiliza las subrutas descompuestas
```
